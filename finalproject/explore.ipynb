```python
import json
import pandas as pd
```

```python
with open("na_ground_truth.json", 'r') as f:
    data = json.load(f)
```

```python
print(data)
```

```python
df = pd.DataFrame(columns=["key-author","title", "year", "name", "id", "label", "org"])

for d in data:
    key_author = d['name']
    pubs = d['pubs']
    for pub in pubs:
        pub.update({'key-author':key_author})
        df = df.append(pub, ignore_index=True)
```

```python
df.to_csv("na_truth_value.csv", index=False)
```
